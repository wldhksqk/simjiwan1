```python
f = open("sin.txt", 'w')
for i in range(len(y)): #len함수 활용 문자열길이 10개~
    data = y[i]
    f.write(data) #이 오류는 str을 넣어야지 np.float64를 넣으면 할 수가 없다라는 뜻.
    f.write('\n')
f.close()
```

**ㆍ불러온 파일의 모든 line 불러오기**

```python
f = open("test.txt", 'r') 
while True: #while 조건문 : 참일 시 계속 한다
    line = f.readline() 
    if not line : break #더 이상 읽을 라인 없을때 → break해라
    print(line)
f.close()
```

# ㆍreadlines 함수

**ㆍwith문 활용하면 f.close()를 생략할 수 있다.**

ㆍwith문 활용하여 파일 쓰기

**(1)**test.txt파일을 w(write)모드로 오픈하는 것을

f라는 이름으로 한다.

**(2)**그 안에서 f라는 이름의 내용은 이것으로 한다!

```python
with open('test2.txt', 'w') as f: #(1)
    f.write("Life is too short, you need python.") #(2)
 
```

**line을 split을 사용해서 단어로 쪼개기.**

```python
f=open('test2.txt', 'r')
line = f.readline()
words = line.split()
print(words)
f.close()
```

```python
with open('test2.txt', 'r') as f: 
  line=f.readline()
  words=line.split()
  for i in range(len(words)): #여기선 i가 숫자, words의 길이를 인덱싱 하고 있기 때문
    print(words[i])
f.close()
    
```

**split한 이후 첫 번째 entry값에 vector값이 저장된다.**

```python
import matplotlib.pyplot as plt

with open("sin.txt", 'r') as f:
  lines = f.readlines()
  y=[] #y라는 변수에
  for line in lines:
    words=line.split(':')
    y.append(eval(words[1])) #eval함수로 문자를 숫자로 변환한 셈
    #y라는 변수에다가 계속 append(추가)한다.

plt.plot(y)
plt.show
f.close()
```
