전역 변수는 선언된 이후 어디서든 접근이 가능!

그렇다면, n이라는 이름의 전역 변수가 선언되었는데, 함수 안에서 변수를 n이라는 이름으로 만들면?

결론)
1. 함수 안에서는 n=0으로 저장 되며, pirnt(lv)를 통하여 1이 출력
2. 함수 밖에서는 print(n)를 하면, 10이 출력

```python
n=10

def func2():
    n=0 #전역 변수 n이 있지만, 함수 안에서 동일한 이름의 변수를 선언하면
        #함수안에서 지역 변수로 n이 만들어지고 사용된다. 휘발성이 적용
    lv=n+3
    print(lv)

func2()
```

**함수 안에서 전역 변수**에 접근하고 싶으면 어떻게 할까?

global 사용

```python
n=20

def func2():
    global n 
    n=5
    lv=n+5
    print(lv)

func2()    
```

**global n**을 통해서, 함수 내의 전역 변수 n에 접근, 따라서 전역 변수 n의 값이 0으로 성공적으로 입력 됨.

```python
print(n)
```

다양한 객체 생성 가능

**객체 마다 함수를 정의할 필요가 없다는 장점!**

```python
class Ageinfo:
    def up_age(self):
        self.age +=5
    def get_age(self):
        return self.age
```

```python
mom = Ageinfo()
uncle = Ageinfo()
me = Ageinfo()

mom.age = 48
uncle.age = 39
me.age = 23

print('엄마의 나이', mom.get_age())
print('삼촌의 나이', uncle.get_age())
print('나의 나이', me.get_age())

print('5년 뒤...')
mom.up_age()
uncle.up_age()
me.up_age()

print('엄마의 나이', mom.get_age())
print('삼촌의 나이', uncle.get_age())
print('나의 나이', me.get_age())
```

```python
import math
import matplotlib.pyplot as plt
class sin_graph:
    def generate_x_y(self): #객체 자기자신(self) 생략해도 됨
        self.x=[] #그냥 빈 배열을 만들어 놓은 거임
        self.y=[]
        self.dx=(self.end-self.st)/(self.n-1)
        for i in range(self.n):
            x_=self.st + self.dx*i #x값을 하고 싶은건데 변수이름 구분하기 그러니까 언더바로 구분
            self.x.append(x_) #append 추가하다 / 반복되는 개수만큼 self.x 및 y 에 붙는다
            self.y.append(math.sin(x_)) 
    def plot(self):
        plt.plot(self.x,self.y)
        plt.show()
```

```python
hi=sin_graph()
hi.st=0
hi.end=2*math.pi
hi.n=200

hi.generate_x_y()
hi.plot()
plt.show()
```

```python
class sin_graph:
    def setting(self, st, ed, n): #말 그대로 내부 변수를 세팅하는 거임
        self.st=st
        self.ed=ed
        self.n=n
    def generate_x_y(self):
        self.x=[]
        self.y=[]
        self.dx=(self.ed-self.st)/(self.n-1)
        for i in range(self.n):
            x_=self.st + self.dx*i
            self.x.append(x_) 
            self.y.append(math.sin(x_)) 
    def plot(self):
        plt.plot(self.x,self.y)
        plt.show()
```

```python
g1=sin_graph()
g1.setting(0, 2*math.pi, 150)
g1.generate_x_y()
g1.plot()
```

```python
class cos_graph:
    def setting(self, st, ed, n):
        self.st=st
        self.ed=ed
        self.n=n
    def generate_x_y(self):
        self.x=[]
        self.y=[]
        self.dx=(self.ed-self.st)/(self.n-1)
        for i in range(self.n):
            x_=self.st + self.dx*i
            self.x.append(x_) 
            self.y.append(math.cos(x_)) 
    def plot(self):
        plt.plot(self.x,self.y)
        plt.show()
```

```python
g2=cos_graph()
g2.setting(0, 16*math.pi, 250)
g2.generate_x_y()
g2.plot()
```

**생성자** 


```python
class Const:
    def __init__(self):
        print("hello~")

h1 = Const()
h2 = Const()
#Const클래스로 객체 h1, h2를 만들자 print("hello~")가 실행된다.
```

**생성자의 특징**

1)클래스 안에 정의

2)이름이 **__init__(self)**

3)객체 생성시 자동으로 호출

4)변수를 초기화 하기 위해서 사용됨

5)클래스 안에 정의를 하지 않아도 에러가 안생기지만 편의성을 위해 정의

```python
#def __init__(self, 매개 변수): ← 매개 변수 초기화 내용
class Const:
    def __init__(self, h1, h2):
        self.h1 = h1
        self.h2 = h2
    def show_data(self):
        print(self.h1, self.h2)

h1 = Const(8,16)
h2 = Const(32,64)
```

```python
h1.show_data()
h2.show_data()
```
