**데이터 생성 함수**

**np.linspace** 

numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)

시작부터 끝의 범위에서 num개를 균일한 간격으로 데이터 생성 및 배열을 만드는 함수

```python
a = np.linspace(2, 20, 10)
print(a)
```

**np.arange**

numpy.arange(start, stop, step, dtype=None)

시작부터 끝 **미만**까지 스텝 간격으로 데이터 생성 후 배열을 만듦

범위내에서 간격을 기준 균등 간격의 배열

요소의 객수가 아닌 **데이터의 간격**을 기준으로 배열 생성

```python
a = np.arange(0, 40, 4, np.int16)
print(a)
```

np.logspace
**굵은 텍스트**
numpy.logspace(start, stop, num=50, endpoint=True, base=10.0, dtype=None)

로그스케일의 linspace 함수

로그스케일로 지정된 범위에서 num개수만큼 균등 간격으로 데이터 생성 후 배열 생성

```python
a = np.logspace(0, 20, 4, np.float64)
print(a)
```

**Sin 함수 그리기**

```python
#X축 생성 : np.linspace를 사용해서, 점을 균일하게 찍을 수 있다
x = np.linspace(0,2*np.pi, 10)
#sin(x)위의 점 생성
y=np.sin(x)
```

```python
#plt.xlabel('X-axis')
#plt.ylabel('Y-axis')
#plt.title('Data') 필요하면 이거 써도 됨
plt.plot(x,y)
plt.show()
#100개의 점으로 생성하면 그래프 곡선처럼 보이지만 지금 이 그래프 처럼 
#사실상 직선으로 이루어진 그래프임
```

**다른 함수 예제**

```python
x = np.linspace(0,2*np.pi, 200)
#x벡터 위의 값 별로 계산이 됨!
y = np.sin(x)+np.cos(x)+10
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x)+cos(x)+10')
plt.show()
```

**두 그래프 같이 그리기(legend)**

plt.legend를 통하여 그래프에 이름을 붙여 줄 수 있다

```python
x = np.linspace(0,2*np.pi, 200)
```

```python
y1 = np.sin(x)+5
y2 = np.sin(x)+np.cos(x)+5
```

```python
plt.plot(x,y1,x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data')
plt.legend(['sin(x)+5', 'sin(x)+cos(x)+5']) #대괄호
plt.show()
```

**Tick 설정하기**

Tick을 이용하여, 그래프의 위치를 표시할 수 있다.

plt.xticks([0, np.pi, 2* np.pi])

plt.yticks([-0, 0, 1])

```python
x = np.linspace(0,2*np.pi, 200)

y1 = np.sin(x)
y2 = np.sin(x)+np.cos(x)

plt.plot(x,y1,x,y2)

plt.xticks([0, np.pi, 2*np.pi])
plt.yticks([-1, 0, 1])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Data')
plt.legend(['sin(x)', 'sin(x)+cos(x)'])

plt.show()
```
